```python
import pandas as pd
import numpy as np
```

```python
# Create data set
sales_data = pd.DataFrame({
    "name": [
        "William", "Emma", "Sofia", "Markus", "Edward", "Thomas", "Ethan",
        np.nan, "Arun", "Anika", "Paulo"
    ],
    "region": [
        np.nan, "North", "East", "South", "West", "West", "South", np.nan,
        "West", "East", "South"
    ],
    "sales": [
        50_000, 52_000, 90_000, np.nan, 42_000, 72_000, 49_000, np.nan, 67_000,
        65_000, 67_000
    ],
    "expenses": [
        42_000, 43_000, np.nan, 44_000, 38_000, 39_000, 42_000, np.nan, 39_000,
        44_000, 45_000
    ]
})
```

```python
from sklearn.impute import SimpleImputer
```

```python
sales_data.style.highlight_null("skyblue")
```

```python
simple_imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
```

```python
simple_imputer
```

```python
fitted_imputer = simple_imputer.fit(sales_data.select_dtypes(exclude="object"))
```

```python
fitted_imputer
```

```python
transformed_imputer = fitted_imputer.transform(sales_data.select_dtypes(exclude="object"))
```

```python
transformed_imputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=3)
transformed = imputer.fit_transform(sales_data.select_dtypes(exclude="object"))
```

```python
transformed
```

```python
sales_data.assign(sales = transformed[:, 0])
```
